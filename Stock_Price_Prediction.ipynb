```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
```

```python
data=pd.read_csv('GOOG.csv',date_parser=True)
data.tail()
```

```python
data_training=data[data['Date']<'2019-01-01'].copy()
data_training
```

```python
data_test=data[data['Date']>='2019-01-01'].copy()
data_test
```

```python
training_data=data_training.drop(['Date','Adj Close'],axis=1)
training_data.head()
```

```python
scaler=MinMaxScaler()
training_data=scaler.fit_transform(training_data)
training_data
```

```python
X_train=[]
Y_train=[]
```

```python
training_data.shape[0]
```

```python
for i in range(60,training_data.shape[0]):
  X_train.append(training_data[i-60:i])
  Y_train.append(training_data[i])
```

```python
X_train,Y_train=np.array(X_train),np.array(Y_train)
```

```python
X_train.shape,Y_train.shape
```

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense,LSTM,Dropout
```

```python
reg=Sequential()
reg.add(LSTM(units=60,activation='relu',return_sequences=True,input_shape=(X_train.shape[1],5)))
reg.add(Dropout(0.2))
reg.add(LSTM(units=60,activation='relu',return_sequences=True))
reg.add(Dropout(0.2))
reg.add(LSTM(units=80,activation='relu',return_sequences=True))
reg.add(Dropout(0.2))
reg.add(LSTM(units=120,activation='relu'))
reg.add(Dropout(0.2))
reg.add(Dense(units=1))
```

```python
reg.summary()
```

```python
reg.compile(optimizer='adam',loss='mean_squared_error')
```

```python
reg.fit(X_train,Y_train,epochs=50,batch_size=32)
```

```python
data_test.head()
```

```python
data_training.tail(60)
```

```python
past_60_days=data_training.tail(60)
```

```python
df=past_60_days.append(data_test,ignore_index=True)
df=df.drop(['Date','Adj Close'],axis=1)
df.head()
```

```python
inputs=scaler.transform(df)
inputs
```

```python
X_test=[]
Y_test=[]
for i in range(60,inputs.shape[0]):
  X_test.append(inputs[i-60:i])
  Y_test.append(inputs[i])
```

```python
X_test,Y_test=np.array(X_test),np.array(Y_test)
X_test.shape,Y_test.shape
```

```python
y_pred=reg.predict(X_test)
```

```python
y_pred
```

```python
scaler.scale_
```

```python
scale=1/8.18605127e-04
scale
```

```python
y_pred=y_pred*scale
Y_test=Y_test*scale
```

```python
plt.figure(figsize=(14,5))
plt.plot(Y_test,color='red',label='Real Google Stock Price')
plt.plot(y_pred,color='blue',label='Predicted Google Stock Price')
plt.title('Google Stock Price Prediction')
plt.xlabel('Time')
plt.ylabel('Google stock Price')
plt.legend()
plt.show()
```
